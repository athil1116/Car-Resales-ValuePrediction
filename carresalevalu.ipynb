```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Datasets/train-data.csv')
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df.drop(['Unnamed: 0','New_Price'],axis=1,inplace=True)
```

```python
df
```

```python
df['Mileage']=df['Mileage'].str.replace('km/kg','')
df['Mileage']=df['Mileage'].str.replace('kmpl','')
```

```python
df['Engine']=df['Engine'].str.replace('CC','')
df['Power']=df['Power'].str.replace('bhp','')
```

```python
df
```

```python
df['Mileage']=df['Mileage'].astype(float)
df['Engine']=df['Engine'].astype(float)
df['Power']=df['Power'].str.replace('null','0')
df['Power']=df['Power'].astype(float)
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df['Engine']=df['Engine'].fillna(df['Engine'].mode()[0])
df['Power']=df['Power'].fillna(df['Power'].mean())
df['Seats']=df['Seats'].fillna(df['Seats'].mode()[0])
df['Mileage']=df['Mileage'].fillna(df['Mileage'].mean())
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Fuel_Type']=le.fit_transform(df['Fuel_Type'])
df['Location']=le.fit_transform(df['Location'])
df['Transmission']=le.fit_transform(df['Transmission'])
df['Owner_Type']=le.fit_transform(df['Owner_Type'])
```

```python
df
```

```python
df.dtypes
```

```python
df['Brand']=df['Name'].apply(lambda x:x.split()[0])
df
```

```python
df['Brand']=le.fit_transform(df['Brand'])
df
```

```python
df.drop(['Name'],axis=1,inplace=True)
```

```python
df
```

```python
X=df.drop(['Price'],axis=1)
X
```

```python
y=df.iloc[:,-2]
y
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_scaled=scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
mlr=LinearRegression()
mlr.fit(X_train,y_train)
y_pred=mlr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score
print('R2_score:',r2_score(y_test,y_pred))
```

```python
import seaborn as sns
sns.regplot(x=df['Power'],y=df['Price'])
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train,y_train)
y_pred_k=knn.predict(X_test)
y_pred_k
```

```python
print('r2_score:',r2_score(y_test,y_pred_k))
```
